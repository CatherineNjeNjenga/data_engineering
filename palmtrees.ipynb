```python
# load libraries
import pandas
import duckdb
```

```python
# load data
palmtrees = pandas.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-03-18/palmtrees.csv', encoding='windows-1252')
palmtrees
```

```python
df = palmtrees.loc[:, :].rename(columns={'max__blade__length_m': 'blade_length', 'max__rachis__length_m': 'rachis_length'})
df = df[df['blade_length'].notna() & df['rachis_length'].notna()]
df = df[(df['acc_genus'] == 'Phoenix') | (df['acc_genus'] == 'Hyphaene') | (df['acc_genus'] == 'Borassus')]
df['acc_species'] = df['acc_species'].str.title()
df_long = df.melt(id_vars = "acc_species", 
                value_vars = ["blade_length", "rachis_length"],
                var_name = "Measurement_Type",
                value_name = "Size")

# df_long[, Size := ifelse(Measurement_Type == "blade_length", -Size, Size)]
def classify_length(size):
    print(size)
    return size if (value == "blade_length") else 0

# df['freq'] = df.apply(lambda x: frequer(x['count']),axis=1)
df_long['Updated'] = df_long.apply(lambda x: classify_length(x))
df_long
```
